```python
import pandas as pd
from termcolor import cprint 
```

# 1. 신용위험 분석 

```python
loan = pd.read_csv('https://media.githubusercontent.com/media/fintech-data/Revolution4/main/loan.csv')
loan.head()
```

```python
loan_status = pd.read_csv("https://raw.githubusercontent.com/fintech-data/Revolution4/main/data/loan_data_description.csv")
```

```python
def description(column_name):
    #description function gives you meaning of the feature 
    return loan_status.loc[loan_status.LoanStatNew == column_name].Description.values[0]

def print_description(column_list):
    #print_description function prints meanings and Null value Percentages of a feature
    for i, column in enumerate(column_list):
        color = 'grey'
        if column == 'loan_status':
            color = 'green'
        try:
            cprint("- {} = {}".format(column, description(column), color))
        except Exception as e:
                   cprint("- {}".format(column), 'red')
            
loan_columns = loan.columns
print_description(loan_columns)
```
